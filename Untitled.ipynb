```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a=2
T=0.5
l1 = l2 = 0
L1 = 10
L2 = 5
N1 = N2 = 100

X = np.linspace(l1, L1, N1)
Y = np.linspace(l2, L2, N2)

def u0(x, y):
    condition = (x >= 4) & (x <= 6) & (y >= 2) & (y <= 3)
    return np.where(condition, 4, 0)

# Gnrer U0 en utilisant meshgrid directement
U0 = u0(*np.meshgrid(X, Y, indexing='ij'))

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create grid coordinates
X, Y = np.meshgrid(X, Y)

# Plot the surface
surf = ax.plot_surface(X, Y, U0, cmap='viridis')
fig.colorbar(surf, label='U0')
ax.set_title('Plot de U0')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('U0')
plt.show()




CFL=0.5
dt1=CFL*dx/a
dt2=CFL*dy/a
dt=min(dt1,dt2)


Un = U0.copy()
Unp1 = np.zeros((N1,N2))
temps=0
while temps <= T:
    for i in range(1,N1-1):
        for j in range(1,N2-1): 
            Unp1[i][j]=Un[i][j]-a*dt/dx*(Un[i][j]-Un[i-1][j])-a*dt/dy*(Un[i][j]-Un[i][j-1])
    for k in range(N2):  
        Unp1[0][k] = Unp1[1][k]
        Unp1[N1-1][k] = Unp1[N1-2][k]
    for l in range(N1):  
        Unp1[l][0] = Unp1[l][1]
        Unp1[l][N2-1] = Unp1[l][N2-2]
    
    temps+=dt
        
# Plot the final solution Un
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create grid coordinates
X, Y = np.meshgrid(np.linspace(l1, L1, N1), np.linspace(l2, L2, N2))

# Plot the surface
surf = ax.plot_surface(X, Y, Un, cmap='viridis')
fig.colorbar(surf, label='Un')
ax.set_title('Plot de Un')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Un')
plt.show()



```
